```python
# Chatbot IA Inteligente tipo ChatGPT con Google Gemini
# Instalar la biblioteca de Google
!pip install -q google-generativeai

import google.generativeai as genai
from datetime import datetime

print("🤖 Configurando el Chatbot IA Inteligente...\n")

# ========================================
# CONFIGURACIÓN DE LA API
# ========================================
print("📝 Para usar este chatbot necesitas una API Key GRATUITA de Google:")
print("   1. Ve a: https://makersuite.google.com/app/apikey")
print("   2. Crea una API key (es gratis)")
print("   3. Cópiala y pégala aquí abajo\n")

# API Key configurada
API_KEY = "AIzaSyCkHMH4Te-khwbXu-vmkMKvDAKWO880zG8"

print("✅ API Key configurada correctamente\n")

# Configurar la API
genai.configure(api_key=API_KEY)

# ========================================
# PERSONALIDAD DEL CHATBOT
# ========================================
SYSTEM_PROMPT = """Eres un asistente de IA amigable, inteligente y útil llamado "AsistenteIA".

Tus características:
- Eres cortés, empático y siempre intentas ayudar
- Puedes responder preguntas sobre cualquier tema
- Das consejos prácticos y útiles
- Explicas conceptos complejos de forma simple
- Mantienes conversaciones naturales
- Hablas principalmente en español
- Cuando te presentas, dices tu nombre con entusiasmo
- Eres creativo y puedes ayudar con tareas, dudas, consejos de vida, programación, etc.

Siempre respondes de forma clara, concisa y amigable."""

# Inicializar el modelo con configuración compatible
try:
    model = genai.GenerativeModel(
        model_name='gemini-pro',
        system_instruction=SYSTEM_PROMPT
    )
    print("✅ Modelo cargado: Gemini Pro\n")
except Exception as e:
    print(f"⚠️ Intentando con modelo alternativo...\n")
    model = genai.GenerativeModel('gemini-pro')

# Iniciar el chat
chat = model.start_chat(history=[])

# Chatbot IA Inteligente - 100% Gratuito y Sin API Key
# Usa modelos de Hugging Face

print("📦 Instalando dependencias...")
!pip install -q transformers torch accelerate sentencepiece

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import warnings
warnings.filterwarnings('ignore')

print("\n🤖 Iniciando Chatbot IA Inteligente...")
print("⏳ Descargando modelo inteligente (puede tomar 2-3 minutos)...\n")

# Usar un modelo más inteligente en español
model_name = "microsoft/DialoGPT-medium"

# Detectar si hay GPU disponible
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"💻 Usando: {device.upper()}")

# Cargar modelo y tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, padding_side='left')
model = AutoModelForCausalLM.from_pretrained(model_name)
model = model.to(device)

# Configurar el pad token
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

print("\n✅ ¡Modelo cargado exitosamente!")
print("\n" + "=" * 70)
print("💬 ASISTENTE IA INTELIGENTE - Tu compañero virtual")
print("=" * 70)
print("🌟 Puedo ayudarte con:")
print("   • Consejos de vida y productividad")
print("   • Responder preguntas")
print("   • Conversaciones amigables")
print("   • Apoyo emocional")
print("   • Ideas creativas")
print("\n💡 Escribe 'salir' para terminar\n")

# Sistema de personalidad mejorado
chat_history_ids = None
conversation_turns = 0

# Mensaje de presentación
print("🤖 AsistenteIA: ¡Hola! Soy tu AsistenteIA, un compañero virtual diseñado")
print("                para ayudarte con lo que necesites. Puedo darte consejos,")
print("                responder preguntas, o simplemente conversar contigo.")
print("                ¿En qué puedo ayudarte hoy?\n")

def generate_response(user_input, chat_history_ids):
    """Genera una respuesta inteligente"""

    # Codificar entrada del usuario
    new_input_ids = tokenizer.encode(
        user_input + tokenizer.eos_token,
        return_tensors='pt'
    ).to(device)

    # Concatenar con historial
    bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1) if chat_history_ids is not None else new_input_ids

    # Generar respuesta con parámetros mejorados
    chat_history_ids = model.generate(
        bot_input_ids,
        max_length=1000,
        pad_token_id=tokenizer.pad_token_id,
        no_repeat_ngram_size=3,
        do_sample=True,
        top_k=100,
        top_p=0.92,
        temperature=0.85,
        repetition_penalty=1.2
    )

    # Decodificar respuesta
    response = tokenizer.decode(
        chat_history_ids[:, bot_input_ids.shape[-1]:][0],
        skip_special_tokens=True
    )

    return response, chat_history_ids

# Respuestas predefinidas inteligentes para casos comunes
RESPUESTAS_INTELIGENTES = {
    'hola': "¡Hola! 😊 Me alegra saludarte. Soy tu AsistenteIA y estoy aquí para ayudarte con lo que necesites. ¿En qué puedo apoyarte hoy?",
    'como te llamas': "Me llamo AsistenteIA. Soy un asistente virtual diseñado para ayudarte, darte consejos y mantener conversaciones interesantes. ¿Y tú, cómo te llamas?",
    'quien eres': "Soy AsistenteIA, tu compañero virtual inteligente. Puedo ayudarte con consejos, responder preguntas, apoyarte emocionalmente o simplemente charlar contigo.",
    'ayuda': "¡Claro! Puedo ayudarte con muchas cosas: dar consejos, responder preguntas, sugerir ideas, apoyarte con problemas, o simplemente conversar. ¿Qué necesitas?",
    'consejo': "Con gusto te doy un consejo. Aquí va uno: La clave del éxito es la constancia. Pequeños pasos cada día te llevan lejos. ¿Hay algo específico sobre lo que quieres un consejo?",
    'triste': "Lamento que te sientas triste. Recuerda que los momentos difíciles son temporales. ¿Quieres hablar sobre lo que te preocupa? Estoy aquí para escucharte.",
    'gracias': "¡De nada! Es un placer ayudarte. Si necesitas algo más, no dudes en preguntarme. 😊"
}

# Bucle principal mejorado
while True:
    try:
        # Obtener entrada del usuario
        user_input = input("Tú: ").strip().lower()

        # Verificar salida
        if user_input in ['salir', 'exit', 'quit', 'adiós', 'adios', 'bye', 'chao']:
            print("\n🤖 AsistenteIA: ¡Fue un placer ayudarte! Recuerda que siempre estaré")
            print("                aquí cuando me necesites. ¡Cuídate mucho y vuelve pronto! 👋✨\n")
            break

        # Verificar entrada vacía
        if not user_input:
            print("🤖 AsistenteIA: Estoy aquí para escucharte. ¿Qué te gustaría saber o conversar?\n")
            continue

        # Buscar respuesta inteligente predefinida
        respuesta_encontrada = False
        for keyword, response in RESPUESTAS_INTELIGENTES.items():
            if keyword in user_input:
                print(f"\n🤖 AsistenteIA: {response}\n")
                respuesta_encontrada = True
                break

        if not respuesta_encontrada:
            # Generar respuesta con el modelo
            response, chat_history_ids = generate_response(user_input, chat_history_ids)

            # Si la respuesta está vacía o es muy corta, dar una respuesta genérica
            if len(response.strip()) < 3:
                response = "Interesante punto de vista. Cuéntame más sobre eso, me gustaría entender mejor tu perspectiva."

            print(f"\n🤖 AsistenteIA: {response}\n")

        conversation_turns += 1

        # Reiniciar historial periódicamente para mantener coherencia
        if conversation_turns > 8:
            chat_history_ids = None
            conversation_turns = 0

    except KeyboardInterrupt:
        print("\n\n🤖 AsistenteIA: ¡Hasta pronto! 👋\n")
        break
    except Exception as e:
        print(f"\n⚠️ Hubo un pequeño error, pero puedo seguir ayudándote.")
        print(f"    Intenta reformular tu pregunta.\n")
        continue

print("✨ ¡Gracias por usar AsistenteIA! Vuelve cuando quieras. 💙")
```
